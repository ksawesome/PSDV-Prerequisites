1. List Operations and Basic Statistics:
    - Create a list of the first 20 natural numbers.
    - Compute the sum of all even numbers in the list.
    - Compute the product of all odd numbers in the list.
    - Calculate the mean, median, and standard deviation of the list.

```python
lst = [int(x) for x in range(1, 21)] # Initialize the list of first 20 natural numbers.

summ = 0 # Initialize variables to store the sum and product.
prod = 1

for i in range(1, 20, 2):
    summ += lst[i] # Calulate the sum of all even numbers in the list.

for i in range(0, 20, 2):
    prod *= lst[i] # Calculate the product of all odd numbers in the list.

mean_lst = sum(lst) / len(lst)

median_lst = (lst[len(lst) // 2 - 1] + lst[len(lst) // 2]) / 2 if len(lst) % 2 == 0 else lst[len(lst) // 2]

def std(data): 
    mean = sum(data) / len(data)  
    variance = sum((x - mean) ** 2 for x in data) / len(data)  
    return variance ** 0.5 

std_lst = std(lst)

print(summ, prod, mean_lst, median_lst, std_lst)
```

3. Fibonacci Sequence:
    - Write a Python function that takes a number n and returns a list of the first n Fibonacci numbers.
    - Plot the Fibonacci sequence using a line graph.

```python
def fibonacci(n): # Function to construct a list of first n fibonacci numbers.
    if n <= 0:
        return []  # Return an empty list for non-positive n
    fib_sequence = [0, 1]  # Initialize the list with the first two Fibonacci numbers
    for _ in range(2, n):
        fib_sequence.append(fib_sequence[-1] + fib_sequence[-2])  # Append the sum of the last two numbers
    return fib_sequence[:n]  # Return the sequence up to the nth number

k = int(input())
fibb = fibonacci(k) # Store the list.

import matplotlib.pyplot as plt

def plot_line_graph(data):
    plt.figure(figsize=(10, 6))  # Set the figure size
    plt.plot(range(1, len(data) + 1), data, marker = 'o', ls = '-', c = 'r')  # Plot the data
    plt.title('Fibonacci Line Graph')  # Set the title of the graph
    plt.xlabel('Index  of Fibonacci Number')  # Label for the x-axis
    plt.ylabel('Value of Fibonacci Number')  # Label for the y-axis
    plt.grid(True)  # Add a grid for better readability
    plt.show()  # Display the graph


plot_line_graph(fibb)
```

6. Palindrome Check:
    - Write a program to check if a string is a palindrome (ignoring spaces and case).
    - Input: "A man a plan a canal Panama"
    - Output: True

```python
def palindrome_check(str):
    lst = [x.lower() for x in str.split(' ')]
    temp = ''.join(lst)
    return True if temp == temp[::-1] else False

print(palindrome_check("A man a plan a canal Panama"))
```

        - Split a string into a list of words.

        - Visualize the word lengths using a bar chart.

```python
s = 'Running through the forest the trees towered above me their leaves whispering in the wind sunlight filtering through creating a dappled pattern on the ground as I moved forward feeling a mix of excitement and a quiet calm'

words = s.lower().split(' ')

words = sorted([word for word in words if len(word) >= 4])

lengths = [len(word) for word in words]

number_of_lengths = [lengths.count(length) for length in set(lengths)]

plt.bar(range(min(lengths), max(lengths) + 1), number_of_lengths)
plt.xlabel('Word Length')
plt.ylabel('Frequency')
plt.title('Frequency of Word Lengths')
plt.show()
```
